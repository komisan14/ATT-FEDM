```python
df = df.astype('string')
```

```python
#Replacing the 'contact agent for price' into null values
def change_price(col):
    price_change = 'Contact agent for price';
    
    df[col][df[col] == price_change] = None;
```

```python
change_price('price')
```

```python
df['price']
```

```python
df['price'] = df['price'].str.replace('₱', '')
df['price'] = df['price'].str.replace(',', '')

df[['price']]
```

```python
df['price'] = df['price'].astype(float)
```

```python
df.drop(['Unnamed: 0'], axis = 1, inplace = True)
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.isna().sum()
```

```python
df['title'] = df['title'].fillna('NA');
df['description'] = df['description'].fillna('NA');
```

```python
sns.boxplot(df['price'])
```

```python
#Fill in price values with median values, the price is not normally distributed 
df['price'].fillna(df['price'].median(),inplace = True);
```

```python
def remove_outliers(df,col):
    q1 = df[col].quantile(0.25);
    q3 = df[col].quantile(0.75);
    iqr = q3 - q1;
    
    max_ = q3 + (1.5*iqr);
    min_ = q1 - (1.5*iqr);
    
    df.loc[df[col] < min_,col] = min_;
    df.loc[df[col] > max_,col] = max_;
```

```python
remove_outliers(df,'price')
```

```python
sns.boxplot(df['price'])
```
