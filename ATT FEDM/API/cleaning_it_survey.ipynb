```python
df_new['Major in College'] = np.where(df_new['Major in College'] == 'Business Analytics Major',1,0)

df_new
```

```python
X = df_new.drop(['Major in College'], axis=1);
y = df_new['Major in College'];

# #importing model selection for training and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=50, train_size=.70);
```

```python
from sklearn.metrics import confusion_matrix;
from sklearn.metrics import RocCurveDisplay;
from sklearn.metrics import classification_report;
```

```python
def evaluate_model(model, X,y):
    X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=45,
                                                 train_size=.80);
    
    model.fit(X_train, y_train);
    
    print(classification_report(y_test,model.predict(X_test),
                                target_names=['Business Analytics Major','Networking']));
    print(confusion_matrix(y_test,model.predict(X_test),
                          labels=[0,1]));
    print('Training score:',model.score(X_train,y_train));
    print('Test score:',model.score(X_test,y_test));
    RocCurveDisplay.from_estimator(model,X_test,y_test);
```
