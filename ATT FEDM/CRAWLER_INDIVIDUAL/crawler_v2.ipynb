```python
from selenium import webdriver
from bs4 import BeautifulSoup;
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service;
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains;
from webdriver_manager.chrome import ChromeDriverManager;
from selenium.webdriver.support.ui import WebDriverWait;
from selenium.webdriver.support import expected_conditions as EC;
import time;
import pandas as pd;
from selenium.common.exceptions import NoSuchElementException


options = Options()
options.add_experimental_option('detach', True);
driver = webdriver.Chrome(service = Service(ChromeDriverManager().install()), options = options);

try:
    
    driver.get("https://shopee.ph/search?keyword=tshirt")
    driver.maximize_window();
    time.sleep(4);
    links = [];
    print('reign')
    tags = driver.find_elements(By.CSS_SELECTOR,'div.shopee-search-item-result__item > a');

    print(tags)

    print('reign')
    for x in tags:
    # print(x.text, "\n");
        link = x.get_attribute('href');
        links.append(link);
    print(links)
    
    
    #Dito me nagkakaeerror po, di ko alam gagamitin, kung class name o css selector. pero ayaw naman both :))
    details = [];
    
    for link in links: 
        driver.get(link);
        WebDriverWait(driver,10).until(EC.presence_of_element_located((By.CSS_SELECTOR,"._44qnta > span")));
        
        # time.sleep(4);
        df = {};
        df['link'] = link;  
        df['productname'] = driver.find_element(By.CSS_SELECTOR,"div._44qnta > span").text.strip();
        df['productprice'] = driver.find_element(By.CLASS_NAME,"pqTWkA").text.strip();
        # df['productRating'] = driver.find_element(By.CSS_SELECTOR,"._046PXf > div").text.strip(); 
        # df['noofRatings'] = driver.find_element(By.CSS_SELECTOR,"._1k47d8").text.strip(); 
        df['noofSold'] = driver.find_element(By.CLASS_NAME,"P3CdcB").text.strip();
        df['shopName'] = driver.find_element(By.CLASS_NAME,"VlDReK").text.strip();
        # df['shipTo'] = driver.find_element(By.CSS_SELECTOR,"span._jqPI6Z > span").text.strip(); 
        # df['shippingFee'] = driver.find_element(By.CLASS_NAME,"IwHyBY").text.strip();
        # df['responseRate'] = driver.find_elements(By.CLASS_NAME,"Xkm22X");
        df['productDescription'] = driver.find_element(By.CLASS_NAME,"irIKAp").text.strip()
        df['time'] = driver.find_elements(By.CLASS_NAME,"shopee-product-rating__time")   
        df['likes'] = driver.find_elements(By.CLASS_NAME,"shopee-product-rating__like-count"); 
        
        details.append(df);

    print(details);
    
   
except NoSuchElementException as exc:
    print(exc)
```
